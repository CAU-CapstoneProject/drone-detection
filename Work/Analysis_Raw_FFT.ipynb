```python
import librosa
import glob
import numpy as np
import pandas as pd
import mglearn
import matplotlib.pyplot as plt
from scipy.fftpack import fft
from sklearn.preprocessing import scale, robust_scale, minmax_scale, maxabs_scale
from sklearn.preprocessing import normalize
```

```python
# Path
files_form = "*.wav"
noise_path = "/Users/junhyuckwoo/capstone/TestFile/ProceedData/noise_44100_1s/" + files_form
wind_path = "/Users/junhyuckwoo/capstone/TestFile/ProceedData/Background_44100_1s/" + files_form
bee_path = "/Users/junhyuckwoo/capstone/TestFile/ProceedData/bee_44100_1s/" + files_form
drone_path = "/Users/junhyuckwoo/capstone/TestFile/ProceedData/phantom2_44100_1s/" + files_form
print("Wind Path: ", wind_path)
print("Bee Path: ", bee_path)
print("Drone Path: ", drone_path)
```

```python
noise_files = glob.glob(noise_path)
wind_files = glob.glob(wind_path)
bee_files = glob.glob(bee_path)
drone_files = glob.glob(drone_path)
files_num = min(len(bee_files), len(wind_files),len(drone_files))
print("# Noise: ", len(noise_files))
print("# Wind: ", len(wind_files))
print("# Bee: ", len(bee_files))
print("# Drone: ", len(drone_files))
print("# Files: ", files_num)    
```

```python
noise_files.sort()
wind_files.sort()
bee_files.sort()
drone_files.sort()
```

```python
raw, sr = librosa.load(noise_files[0], sr=44100)
noise = maxabs_scale(fft(maxabs_scale(raw)))
raw, sr = librosa.load(wind_files[0], sr=44100)
wind = maxabs_scale(fft(maxabs_scale(raw)))
raw, sr = librosa.load(bee_files[0], sr=44100)
bee =  maxabs_scale(fft(maxabs_scale(raw)))
raw, sr = librosa.load(drone_files[0], sr=44100)
drone = maxabs_scale(fft(maxabs_scale(raw)))

print("Wind shape: ", wind.shape)
print("Bee shape:  ", bee.shape)
print("Drone shape: ", drone.shape)
```

```python
fig1 = plt.figure()
```

```python
#files_num = 10
noise_buf = []
wind_buf = []
bee_buf = []
drone_buf = []
for index in range(1, files_num):
    ### Noise
    raw, sr = librosa.load(noise_files[index], sr=44100)
    fft_noise = maxabs_scale(fft(maxabs_scale(raw)))
    noise = np.concatenate((noise, fft_noise))
    print("Wind shape: ", noise.shape)
    
    ### Wind
    raw, sr = librosa.load(wind_files[index], sr=44100)
    fft_wind = maxabs_scale(fft(maxabs_scale(raw)))
    wind = np.concatenate((wind, fft_wind))
    print("Wind shape: ", wind.shape)
    
    ### Bee
    raw, sr = librosa.load(bee_files[index], sr=44100)
    fft_bee = maxabs_scale( fft(maxabs_scale(raw)))
    bee = np.concatenate((bee, fft_bee))
    print("Bee shape:  ", bee.shape)
    
    ### Drone
    raw, sr = librosa.load(drone_files[index], sr=44100)
    fft_drone =  maxabs_scale(fft(maxabs_scale(raw)))
    bee = np.concatenate((bee, fft_drone))
    print("Bee shape:  ", drone.shape)
    
    ## Chcek index
    buf = [0, 0, 0, 0]
    for id in range(int(sr/2)):
        if abs(fft_noise[id]) >= 0.2:
            buf[0] += id
        if abs(fft_wind[id]) >= 0.2:
            buf[1] += id
        if abs(fft_bee[id]) >= 0.2:
            buf[2] += id
        if abs(fft_drone[id]) >= 0.2:
            buf[3] += id
    noise_buf.append(buf[0])
    wind_buf.append(buf[1])
    bee_buf.append(buf[2])
    drone_buf.append(buf[3])
       
    print(str(index) + " bee is loaded")
    fig1 = plt.plot(fft_drone[:int(sr/2)])
    #fig1 = plt.plot(fft_bee[:int(sr/2)])
    #fig1 = plt.plot(fft_wind[:int(sr/2)])
```
